**Drop the unwanted columns**

```python
df.drop(columns=['overview','homepage','tagline','poster_path','backdrop_path'],axis=1,inplace=True)
```

```python
df.drop(columns=['created_by'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['adult']=df['adult'].astype(object)
df['in_production']=df['in_production'].astype(object)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
Le=LabelEncoder()
```

```python
df['adult']=Le.fit_transform(df['adult'])
df['in_production']=Le.fit_transform(df['in_production'])
```

```python
df.head()
```

**Changing the datatypes**

```python
df['first_air_date']=pd.to_datetime(df['first_air_date'])
df['last_air_date']=pd.to_datetime(df['last_air_date'])
```

```python
df['adult']=df['adult'].astype(int)
df['in_production']=df['in_production'].astype(int)
```

**Replace null values by statistical method**

```python
df.info()
```

```python
df['production_countries']=df['production_countries'].fillna(df['production_countries'].mode)
df['production_companies']=df['production_companies'].fillna(df['production_companies'].mode)
df['spoken_languages']=df['spoken_languages'].fillna(df['spoken_languages'].mode)
df['origin_country']=df['origin_country'].fillna(df['origin_country'].mode)
df['networks']=df['networks'].fillna(df['networks'].mode)
df['languages']=df['languages'].fillna(df['languages'].mode)
df['genres']=df['genres'].fillna(df['genres'].mode)
df['name']=df['name'].fillna(df['name'].mode)
df['original_name']=df['original_name'].fillna(df['original_name'].mode)
df['first_air_date']=df['first_air_date'].fillna(df['first_air_date'].mode)
df['last_air_date']=df['last_air_date'].fillna(df['last_air_date'].mode)

```
